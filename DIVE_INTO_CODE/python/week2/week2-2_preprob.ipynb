# ダミーデータの作成

【問題1】乱数の作成

```python
import numpy as np
np.random.seed(0)
```

```python
mean = [-3, 0]
mean = np.array(mean)
conv = [1.0, 0.8, 0.8, 1.0]
conv = np.array(conv).reshape(2,2)
rand = np.random.multivariate_normal(mean, conv, size=500)
rand.shape
```

【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(rand[:,0], rand[:,1])
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('random_value')
plt.show()
```

【問題3】ヒストグラムによる可視化

```python
plt.title('histogram of x1')
plt.xlabel('x1')
plt.ylabel('frequency')
plt.xlim(right=10, left=-10)
plt.hist(rand[:,0])
plt.show()
```

```python
plt.title('histogram of x2')
plt.xlabel('x2')
plt.ylabel('frequency')
plt.xlim(right=10, left=-10)
plt.hist(rand[:,1])
plt.show()
```

問題4】データの追加

```python
mean2 = [0, -3]
mean = np.array(mean)
rand2 = np.random.multivariate_normal(mean2, conv, size=500)

plt.scatter(rand[:,0], rand[:,1], label='0')
plt.scatter(rand2[:,0], rand2[:,1], label='1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('random_value')
plt.legend()
plt.show()
```

【問題5】データの結合

```python
merge = np.concatenate((rand, rand2), axis=0)
merge.shape
```

【問題6】ラベル付け

```python
rand_label = np.zeros(500)
rand2_label = np.ones(500)
label_merge = np.concatenate((rand_label, rand2_label), axis=0)
label_merge = label_merge.reshape(1000,1)
label_maege_marge = np.append(merge, label_merge, axis=1)
label_maege_marge
```
