# 曽呂利新左衛門問題

【課題】100日後の米粒の数

```python
import matplotlib.pyplot as plt
%matplotlib inline

take_rice = 1
tmp = take_rice
total_rice = [take_rice]
days = 100
list_rice = [take_rice]

for _ in range(days):
    list_rice.append(list_rice[-1]*2)
    tmp += list_rice[-1]
    total_rice.append(tmp)

plt.title('take_rice_curv')
plt.xlabel("number of folds")
plt.ylabel("take_rice[seeds]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(list_rice, label='per day rice')
plt.plot(total_rice, label='total rice')
plt.legend()
plt.show()
```

【課題】100日後以外の米粒の数

```python
import numpy as np
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    take_rice = 1
    tmp = take_rice
    total_rice = [take_rice]
    list_n_rice = [take_rice]
    for d in range(day):
        list_n_rice.append(list_n_rice[-1]*2)
        tmp += list_n_rice[-1]
        total_rice.append(tmp)
    return list_n_rice, total_rice
list_n_rice, total_rice = compute_sorori_shinzaemon(day=10)
print(total_rice)
print(list_n_rice)
plt.title('take_rice_curv')
plt.xlabel("number of folds")
plt.ylabel("take_rice[seeds]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(total_rice, label='total rice')
plt.plot(list_n_rice, label='per day rice')
plt.legend()
plt.show()
```

【課題】何人が何日生活できる米か

```python
# 米の重量：白米の場合→50粒で1g, 玄米の場合→25粒で1g 今回は白米を採用
# 参考サイト→https://ameblo.jp/fenegi/entry-10598490252.html
# 設定を戦国時代の1人が1日あたりで消費する米の量にした
# 理由：現代人は食の選択肢が多すぎる為、米中心の食生活の時代背景にして問題を単純化する為
# DIC同期受講生等を戦国時代の足軽武将と過程した場合、１人あたりの米消費量は約750g/日(米5合相当)
# 参考サイト→https://emuzu-2.cocolog-nifty.com/blog/2014/02/15-a6dc.html

def ToAliveRice(total_rice, people):
    rice_weight = total_rice / 50 * 1
    day = rice_weight / (750 * people)
    return day
_, total_rice = compute_sorori_shinzaemon(day=115)
day = ToAliveRice(total_rice=total_rice[-1], people=14)
print('{}日生存可能'.format(day))
```
