```python
import matplotlib.pyplot as plt
%matplotlib inline


take_rice = 1
days = 100
list_rice = [take_rice]
# list_rice = [list_rice[-1] * 2 for _ in range(days)]
for _ in range(days):
    list_rice.append(list_rice[-1]*2)
# print(list_rice)
plt.title('take_rice_curv')
plt.xlabel("number of folds")
plt.ylabel("take_rice[seeds]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(list_rice)
plt.show()
```

```python
import numpy as np
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    take_rice = 1
    tmp = take_rice
    total_rice = [take_rice]
    list_n_rice = [take_rice]
    for d in range(day):
        list_n_rice.append(list_n_rice[-1]*2)
        tmp += list_n_rice[-1]*2
        total_rice.append(tmp)
    return list_n_rice, total_rice
list_n_rice, total_rice = compute_sorori_shinzaemon(day=10)
print(total_rice)
print(list_n_rice)
plt.title('take_rice_curv')
plt.xlabel("number of folds")
plt.ylabel("take_rice[seeds]")
plt.tick_params(labelsize=20) # 軸の値に関する設定を行う
plt.plot(total_rice)
plt.plot(list_n_rice)
plt.show()
```

```python
def aaa(total_rice, people):
    rice_weight = total_rice / 50 * 1
    year = rice_weight / (400 * people) / 365
    month = (year - int(year)) * 30
    day = (month - int(month)) * 24
    return year, month, day
_, total_rice = compute_sorori_shinzaemon(day=115)
year, month, day = aaa(total_rice=total_rice[-1], people=14)
print(year, month, day)
print('{0}年{1}月{2}日生存可能'.format(year, month, day))
```
