```python
# 栗まんじゅう=100cm^3
# 太陽系=100億光年の球（半径r=100億光年）= 10000000000 * 9.461e+15(m)
# V = 4/3*pi*r^3
import matplotlib.pyplot as plt
%matplotlib inline


MANJU = 0.0001
universe = 3 * 10**80

def calc_full_volume(volume=universe):
    count = 0
    manju_volume_list = [MANJU]
    manju_volume = MANJU
    while manju_volume < volume:
        manju_volume *= 2**12
        manju_volume_list.append(manju_volume)
        count += 1
        
    return count, manju_volume_list

time, manju_volume_list = calc_full_volume(volume=universe)
print('約{}時間後に宇宙が埋め尽くされます'.format(time))

plt.title('manju_vs_universe')
plt.xlabel('hour[h]')
plt.ylabel('manju_volume')
plt.plot(manju_volume_list)
plt.show()
```

```python
time, manju_volume_list = calc_full_volume(volume=1240000)
print('約{}時間後に東京ドームが埋め尽くされます'.format(time))

plt.title('manju_vs_Tokyodome')
plt.xlabel('hour[h]')
plt.ylabel('manju_volume')
plt.plot(manju_volume_list)
plt.show()
```
