# アヤメの分析

【問題1】データの取得

```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
iris
```

```python
X = pd.DataFrame(iris.data, columns=(['sepal_length', 'sepal_width', 'petal_lenght', 'petal_width']))
y = pd.DataFrame(iris.target, columns=(['Species']))
X.head()
```

```python
y.head()
```

【問題2】データの結合

```python
df = pd.concat([X, y], axis=1)
df.head()
```

【問題3】データの確認

```python
# 先頭から4番目までのサンプルを表示させ、各特徴量がどういったものかを見る。（数値データか文字データかなど）
print(df.dtypes)
```

```python
df.head()
```

```python
# ラベルごとのサンプル数の合計を出力する。
df.groupby('Species').count()
```

```python
# 特徴量に欠損値があるかを確認する。
df.isnull().sum()
```

```python
# 特徴量の平均値と標準偏差、四分位数を一度に表示させる。
df.describe()
```

【問題5】必要なデータを抽出

```python
# sepal_widthの列を３つの異なる方法で抽出する
sepal_width = df.sepal_width
print(sepal_width)

sepal_width2 = df.iloc[:, 1]
print(sepal_width2)

sepal_width3 = df.loc[:, 'sepal_width']
print(sepal_width3)
```

```python
# 50番目から99番目までのデータを抽出する
fiftytoninety = df[50:100]
print(fiftytoninety)
```

```python
# petal_lengthの列の50番目から99番目までのデータを抽出する
petal_length_50_99 = df.loc[50:99, 'petal_lenght']
print(petal_length_50_99)
```

```python
# petal_widthの値が0.2のデータを抽出する
petal_width_0_2 = df[df.petal_width == 0.2]
print(petal_width_0_2)
```

.locや.ilocを使うことになります。この２つはどう違うのかについて説明してください。</p>
<br>.loc:行ラベル、 列ラベルを用いてデータを参照する<br>
.iloc:行の番号、列の番号を用いてデータを参照する</p>

【問題6】図の作成

```python
# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
import matplotlib.pyplot as plt
%matplotlib inline

a = df.groupby('Species').count()
b = a.iloc[:, 1]
c = list(b)
target = iris.target_names

plt.pie(c, labels=target, autopct='%1.1f%%')
plt.show()
```

```python
# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
df.boxplot(by='Species')
```

```python
# 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する
import seaborn as sns

sns.catplot(x="Species", y="petal_lenght", kind="violin",data=df)
sns.catplot(x="Species", y="petal_width", kind="violin",data=df)
sns.catplot(x="Species", y="sepal_length", kind="violin",data=df)
sns.catplot(x="Species", y="sepal_width", kind="violin",data=df)
```

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。<br>
箱ひげ図：4分位点や、最大値、最小値、外れ値と言った統計量を直感的に表示することができるが、最大最小値と第一、第三分位点の間にデータがないような印象をもたれてしまう。<br>
バイオリン図：あるカテゴリーのある値に、どのくらいのデータが分布しているかを表示することができる。

【問題7】特徴量間の関係の確認

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る(６つ)
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.scatter(df['sepal_length'], df['sepal_width'], c='r')
plt.show()

plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.scatter(df['sepal_length'], df['petal_lenght'], c='b')
plt.show()

plt.xlabel('sepal_length')
plt.ylabel('petal_width')
plt.scatter(df['sepal_length'], df['petal_width'], c='g')
plt.show()

plt.xlabel('sepal_width')
plt.ylabel('petal_length')
plt.scatter(df['sepal_width'], df['petal_lenght'], c='orange')
plt.show()

plt.xlabel('sepal_width')
plt.ylabel('petal_width')
plt.scatter(df['sepal_width'], df['petal_width'], c='black')
plt.show()

plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.scatter(df['petal_lenght'], df['petal_width'], c='yellow')
plt.show()
```

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る(１つ)
sns.pairplot(df)
```

```python
# ４つの特徴量に関する相関係数行列を作る
corr = df.corr()
```

```python
# 相関係数行列のヒートマップを作る(１つ)
sns.heatmap(corr)
```

【問題8】可視化したグラフや表の結果の説明

散布図行列から、setosa種は他2種類と比較してがく片・花弁の長さ、幅共に小さい傾向にある。かつ、データの分布としても何かしらの直線を引くことで分類可能でありそうなことから、setosa=小さい　という学習をさせることで効果的な推論が可能となると推測できる。<br>
一方でversicolor種、verginica種は、散布図行列のがく片、花弁の長さ・幅の分布状況からは分類が困難に思える。<br>
しかし、バイオリンプロット・箱ひげ図から、versicolor種よりverginica種の方ががく片・花弁共に大きい傾向にあることが窺える。<br>
それが、花弁の長さ・幅両方で特に顕著に現れており、また値のばらつきもその他特徴と比較して小さい。<br>
このことから、種類ごとの花弁の長さ・幅の標準偏差を用いる事で、versicolor種、verginica種も分類することができるのではないかと考える。<br>
