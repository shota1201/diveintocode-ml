```python
def scratch_train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0):
    """
    検証データを分割する。
    Parameters
    ----------
    X : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    y : 次の形のndarray, shape (n_samples, )
      正解値
    train_size : float (0<train_size<1)
      何割をtrainとするか指定
    Returns
    ----------
    X_train : 次の形のndarray, shape (n_samples, n_features)
      訓練データ
    X_test : 次の形のndarray, shape (n_samples, n_features)
      検証データ
    y_train : 次の形のndarray, shape (n_samples, )
      訓練データの正解値
    y_test : 次の形のndarray, shape (n_samples, )
      検証データの正解値
    """
    train_slice = 1 - test_size
    test_slice = 1 - train_size
    
#     if (type(X) is pd.core.frame.DataFrame):
#         X_train = X.loc[:len(X) * train_slice - 1, :]
#         X_test = X.loc[len(X) * train_slice:, :]
        
#     elif (type(X) == np.ndarray) and (X.ndim == 1):
#         array_train_size = int(X.shape[0] * train_size)
#         array_test_size = int(X.shape[0] * test_size)
#         X_train = X[:array_train_size]
#         X_test = X[array_train_size:]

#     elif (type(X) == np.ndarray) and (X.ndim >= 2):
#         array_train_size = int(X.shape[0] * train_size)
#         array_test_size = int(X.shape[0] * test_size)
#         X_train = X[:array_train_size, :]
#         X_test = X[array_train_size:, :]
            
#     return X_train, X_test
    
    if (type(X) is pd.core.frame.DataFrame):
        X = X.sample(frac = 1, random_state = random_state).reset_index(drop=True)
            
        X_train = X.loc[:len(X) * train_slice - 1, :]
        X_test = X.loc[len(X) * train_slice:, :]

    elif (type(X) == np.ndarray) and (X.ndim == 1):
        if random_state == 0:
            random.seed(random_state)
            random.shuffle(X)
        else:
            random.seed(random_state)
            random.shuffle(X)
            
        array_train_size = int(X.shape[0] * train_slice)
        array_test_size = int(X.shape[0] * test_slice)
        X_train = X[:array_train_size]
        X_test = X[array_train_size:]

    elif (type(X) == np.ndarray) and (X.ndim >= 2):
        if random_state == 0:
            random.seed(random_state)
            random.shuffle(X)
        else:
            random.seed(random_state)
            random.shuffle(X)
            
        array_train_size = int(X.shape[0] * train_slice)
        array_test_size = int(X.shape[0] * test_slice)
        X_train = X[:array_train_size, :]
        X_test = X[array_train_size:, :]


    if (type(y) is pd.core.frame.DataFrame):
        y = y.sample(frac = 1, random_state = random_state).reset_index(drop=True)
            
        y_train = y.loc[:len(y) * train_slice - 1, :]
        y_test = y.loc[len(y) * train_slice:, :]

    elif (type(y) == np.ndarray) and (y.ndim == 1):
        if random_state == 0:
            random.seed(random_state)
            random.shuffle(y)
        else:
            random.seed(random_state)
            random.shuffle(y)
        array_train_size = int(y.shape[0] * train_slice)
        array_test_size = int(y.shape[0] * test_slice)
        y_train = y[:array_train_size]
        y_test = y[array_train_size:]

    elif (type(y) == np.ndarray) and (y.ndim >= 2):
        if random_state == 0:
            random.seed(random_state)
            random.shuffle(y)
        else:
            random.seed(random_state)
            random.shuffle(y)
        array_train_size = int(y.shape[0] * train_slice)
        array_test_size = int(y.shape[0] * test_slice)
        y_train = y[:array_train_size, :]
        y_test = y[array_train_size:, :]

    return X_train, X_test, y_train, y_test
```

【問題2】 分類問題を解くコードの作成

```python
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
```

```python
# アイリスデータセット
iris = load_iris()
iris
```

```python
X = pd.DataFrame(iris.data, columns=(['sepal_length', 'sepal_width', 'petal_lenght', 'petal_width']))
y = pd.DataFrame(iris.target, columns=(['Species']))
new_X = X.loc[50:, :]
new_y = y.loc[50:, :]
new_y
```

```python
X_train, X_test, y_train, y_test = scratch_train_test_split(new_X, new_y, test_size=0.2)
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

ss = StandardScaler()
X_train_norm = ss.fit_transform(X_train)
X_test_norm = ss.transform(X_test)

SGD = SGDClassifier(loss='log')
SVC = SVC()
DT = DecisionTreeClassifier()

SGD.fit(X_train_norm, y_train)
SVC.fit(X_train_norm, y_train)
DT.fit(X_train_norm, y_train)

SGD_result = SGD.predict(X_test_norm)
SVC_result = SVC.predict(X_test_norm)
DT_result = DT.predict(X_test_norm)


sgd_accuracy = metrics.accuracy_score(y_test, SGD_result)
print('SGD正解率: {}%'.format(sgd_accuracy*100))

sgd_precision = metrics.precision_score(y_test, SGD_result)
print('SGD適合率: {}%'.format(sgd_precision*100))

sgd_recall = metrics.recall_score(y_test, SGD_result)
print('SGD再現率: {}%'.format(sgd_recall*100))

sgd_f_value = metrics.f1_score(y_test, SGD_result)
print('SGDF値: {}'.format(sgd_f_value))

print('#'*50)

svm_accuracy = metrics.accuracy_score(y_test, SVC_result)
print('SVM正解率: {}%'.format(svm_accuracy*100))

svm_precision = metrics.precision_score(y_test, SVC_result)
print('SVM適合率: {}%'.format(svm_precision*100))

svm_recall = metrics.recall_score(y_test, SVC_result)
print('SVM再現率: {}%'.format(svm_recall*100))

svm_f_value = metrics.f1_score(y_test, SVC_result)
print('SVMF値: {}'.format(svm_f_value))

print('#'*50)

dt_accuracy = metrics.accuracy_score(y_test, DT_result)
print('DT正解率: {}%'.format(dt_accuracy*100))

dt_precision = metrics.precision_score(y_test, DT_result)
print('DT適合率: {}%'.format(dt_precision*100))

dt_recall = metrics.recall_score(y_test, DT_result)
print('DT再現率: {}%'.format(dt_recall*100))

dt_f_value = metrics.f1_score(y_test, DT_result)
print('DTF値: {}'.format(dt_f_value))
```

```python
X_train, X_test, y_train, y_test = scratch_train_test_split(new_X, new_y, test_size=0.2)
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train).reshape(-1)
y_test = np.array(y_test).reshape(-1)

SGD = SGDRegressor()
SGD.fit(X_train, y_train)
SGD_result = SGD.predict(X_test)


sgd_MSE = metrics.mean_squared_error(y_test, SGD_result)
print('平均二乗誤差: {}'.format(sgd_MSE))
```
